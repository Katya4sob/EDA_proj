```python
print("Number of different countries: ", ds.nunique()['Country'])
print("Data belongs to every country:\n", ds.groupby("Country").size())
ds_g=ds.groupby("Country").size()
# for loop with 4 cycles - number of different countries in the data
for key, val in ds_g.items():
    print(key, ":\n") 
    # First plot
    fig = plt.figure(figsize =(15, 10))
    # X, Y axises and plot
    x=ds[ds["Country"]==key].loc[:,'Year']
    y=ds[ds["Country"]==key].loc[:,'Twin_deliveries']
    plt.scatter(x,y,label='deliveries')
    y=ds[ds["Country"]==key].loc[:, 'Twin_children']
    plt.scatter(x,y,label='children')
    # Add title and labels
    plt.title(f"Twin deliveries and children in {key}", fontsize=30)
    plt.xlabel("X - Year", fontsize=20)
    plt.ylabel("Y - Number of deliveries, children", fontsize=20)
    plt.tick_params(axis='both', which='major', labelsize=14)
    # add legend to plot
    plt.legend(fontsize=13)
    plt.grid()
    plt.show() 
    # Second plot
    fig = plt.figure(figsize =(15, 10))
    # Y axises and plot
    y=ds[ds["Country"]==key].loc[:,'Triplet_deliveries']
    plt.scatter(x,y,label='deliveries')
    y=ds[ds["Country"]==key].loc[:,'Triplet_children']
    plt.scatter(x,y,label='children')
    # Add title and labels
    plt.title(f"Triplet deliveries and children in {key}", fontsize=30)
    plt.xlabel("X - Year", fontsize=20)
    plt.ylabel("Y - Number of deliveries, children", fontsize=20)
    plt.tick_params(axis='both', which='major', labelsize=14)
    # add legend to plot
    plt.legend(fontsize=13)
    plt.grid()
    plt.show()
    # Third plot
    fig = plt.figure(figsize =(15, 10))
    # Y axises and plot
    y=ds[ds["Country"]==key].loc[:,'Quadruplet_plus_deliveries']
    plt.scatter(x,y,label='deliveries')
    y=ds[ds["Country"]==key].loc[:,'Quadruplet_plus_children']
    plt.scatter(x,y,label='children')
    # Add title and labels
    plt.title(f"Quadruplet plus deliveries and children in {key}", fontsize=30)
    plt.xlabel("X - Year", fontsize=20)
    plt.ylabel("Y - Number of deliveries, children", fontsize=20)
    plt.tick_params(axis='both', which='major', labelsize=14)
    # add legend to plot
    plt.legend(fontsize=13)
    plt.grid()
    plt.show()
    
```

```python
# same for loop: for loop with 4 cycles - number of different countries in the data
# Create the figure and axes for twinning rate
plt.figure(figsize=(20, 10))
ax = plt.gca()
for key, val in ds_g.items():
    print(key,val, ":\n")  
    # Plot each histogram
    data=ds[ds['Country']==key].loc[:,'Twinning_rate']
    # Filter out NaN values
    filtered_data = data[~np.isnan(data)]
    filtered_data.hist(color=((val%10)*0.1,((100-val%100)%10)*0.1,(val/100-0.8),0.2), label=f'twinning_rate_{key}')
    filtered_data.hist(histtype='step', linewidth=6, label=f'twinning_rate_{key}')
# Add title and axis's labels
plt.title("Twinning rate", fontsize=30)
plt.xlabel("X - Twinning rate", fontsize=20)
plt.ylabel("Y - Amount", fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)   
# Add legend to plot
plt.legend(fontsize=15)
plt.show()

# Create the figure and axes for multiple rate
plt.figure(figsize=(20, 10))
ax = plt.gca()
for key, val in ds_g.items():
    print(key,val, ":\n") 
    # Plot each histogram
    data=ds[ds['Country']==key].loc[:,'Multiple_rate']
    # Filter out NaN values
    filtered_data = data[~np.isnan(data)]
    filtered_data.hist(color=((val%10)*0.1,((100-val%100)%10)*0.1,(val/100-0.8),0.2), label=f'multiple_rate_{key}')
    filtered_data.hist(histtype='step', linewidth=6, label=f'multiple_rate_{key}')
# Add title and axis's labels
plt.title("Multiple rate", fontsize=30)
plt.xlabel("X - Multiple rate", fontsize=20)
plt.ylabel("Y - Amount", fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)
# Add legend to plot
plt.legend(fontsize=15)
plt.show()

# Create the figure and axes for twinning rate
plt.figure(figsize=(20, 10))
ax = plt.gca()
for key, val in ds_g.items():
    print(key, ":\n")
    x=ds[ds["Country"]==key].loc[:,'Year']
    y=ds[ds["Country"]==key].loc[:,'Twinning_rate']
    # Plot each scatter plot
    ax.scatter(x,y, label=f"twinning_rate_{key}")
# Add title and labels
plt.title("Twinning rate", fontsize=30)
plt.xlabel("X - Year", fontsize=20)
plt.ylabel("Y - Twinning rate", fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)
# Add legend and grid to plot
plt.legend(fontsize=15)
plt.grid()
plt.show()

# Create the figure and axes for multiple rate        
plt.figure(figsize=(20, 10))
ax = plt.gca()
for key, val in ds_g.items():
    print(key, ":\n")
    x=ds[ds["Country"]==key].loc[:,'Year']
    y=ds[ds["Country"]==key].loc[:,'Multiple_rate']
    # Plot each scatter plot
    ax.scatter(x,y, label=f"multiple_rate_{key}")
# Add title and labels
plt.title("Multiple rate", fontsize=30)
plt.xlabel("X - Year", fontsize=20)
plt.ylabel("Y - Multiple rate", fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=20)
# Add legend and grid to plot
plt.legend(fontsize=15)
plt.grid()
plt.show()    
```

```python
# same for loop: for loop with 4 cycles - number of different countries in the data
# Create the figure and axes
for key, val in ds_g.items():
    print("\n")
    print(key,val, 'rows')  
    df=ds[ds['Country']==key].loc[:,'Quadruplet_plus_deliveries'] # rows of specific country in 'Quadruplet_plus_deliveries' column
    max_quadruplet_plus_deliveries=df.max() # max value of 'Quadruplet_plus_deliveries'
    max_indices = df[df[:] == max_quadruplet_plus_deliveries].index.tolist() # indeces of max values
    years=ds.loc[max_indices,'Year'] # years with max value of Quadruplet_plus_deliveries
    print(f"max in {key}=", max_quadruplet_plus_deliveries, "and indices of max=", *max_indices)
    print("in the years:", *years.values, sep = "\n")
    
    # additional information about quadruplet children, because somethimes it is not enough information from deliveries only
    df=ds[ds['Country']==key].loc[:,'Quadruplet_plus_children'] # rows of specific country in 'Quadruplet_plus_deliveries' column
    max_quadruplet_plus_children=df.max() # max value of 'Quadruplet_plus_children'
    #print("\n")
    print("max_quadruplet_plus_children=", max_quadruplet_plus_children)
    max_indices_ch = df[df[:] == max_quadruplet_plus_children].index.tolist() # indeces of max values
    # check if there are same indices, if so we don't need this additional info
    i=0
    j=0
    flag=0
    for i in range(len(max_indices)):
        if (flag==0):
            for j in range(len(max_indices_ch)):
                if(max_indices[i]==max_indices_ch[j]):
                    flag=1
                    break
        else:
            break 
    if (flag):  
        continue
    else:    
        years=ds.loc[max_indices_ch,'Year'] # years with max value of Quadruplet_plus_children
        print(f"Info from children, max in {key}=", max_quadruplet_plus_children//4, "and indices of max=", *max_indices_ch)
        print("in the years:", *years.values, sep = "\n")
        print("\n")
```
